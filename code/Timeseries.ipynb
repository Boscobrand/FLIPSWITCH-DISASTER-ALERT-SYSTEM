```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
july22=pd.read_csv('test_data_July22nd.csv')
july23=pd.read_csv('test_data_July23rd.csv')
july24=pd.read_csv('test_data_July24th.csv')
july25=pd.read_csv('test_data_July25th.csv')
july26=pd.read_csv('test_data_July26th.csv')
july27=pd.read_csv('test_data_July27th.csv')
july28=pd.read_csv('test_data_July28th.csv')
july29=pd.read_csv('test_data_July29th.csv')
july30=pd.read_csv('test_data_July30th.csv')
```

```python
test_data=pd.concat([july22,july23,july24,july25,july26,july27,july28,july29,july30])
```

```python
test_data.head()
```

```python
test_data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
test_data['created_at']=pd.to_datetime(test_data['created_at'])
test_data.set_index('created_at',inplace=True)
test_data.sort_index(inplace=True)
```

```python
hours=test_data.resample('4H').sum()
hours
```

```python
plt.figure(figsize=(12,9))
plt.plot(hours['preds'],color='blue');
```

```python
plt.figure(figsize=(16,9))
plt.plot(hours['label'],color='orange')
plt.plot(hours['preds'],color='blue');
```

```python
days=test_data.resample('D').sum()
days
```

```python
plt.figure(figsize=(12,9))
plt.plot(days['label'],color='orange')
plt.plot(days['preds'],color='blue');
```

```python
plt.figure(figsize=(12,9))
plt.plot(days['preds'],color='blue');
```
